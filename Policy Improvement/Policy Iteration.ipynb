```python
Q=np.zeros((4,6,6))
V=np.zeros((6,1))
V_prev=np.ones((6,1))
episilon=0.001
gamma=0.99

log=[]
while ((V-V_prev)**2).sum()**0.5>episilon:
    Q=R_tr+gamma*np.matmul(P_tr,V)
    V_prev=V.copy()
    V=np.max(Q,axis=0)
    pi=np.argmax(Q,axis=0)
    log.append((pi.flatten(),V.sum()))
```

```python
Q=np.zeros((4,6,6))
V=np.zeros((6,1))
V_prev=np.ones((6,1))
episilon=0.001
gamma=0.99

log=[]
while ((V-V_prev)**2).sum()**0.5>episilon:
    Q=R_tr+gamma*np.matmul(P_tr,V)
    pi=((np.e**Q)/np.sum(np.e**Q,axis=0))
    V_prev=V.copy()
    V=np.multiply(((np.e**Q)/np.sum(np.e**Q,axis=0)),Q).sum(axis=0)
    log.append((pi.reshape(4,6).transpose(1,0),V.sum()))
```
